```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set up aesthetic defaults
sns.set_theme(style="whitegrid")
```

## 1. Load Macroeconomic Data (FRED)

```python
df_fred = pd.read_csv("fred_macro.csv", parse_dates=["DATE"])
df_fred.set_index("DATE", inplace=True)
df_fred.head()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_fred, x=df_fred.index, y="UNRATE", label="Unemployment Rate")
plt.title("FRED Unemployment Rate")
plt.show()
```

## 2. Dynamic Causal Inference
Let's import `wage.csv` and run a quick Ordinary Least Squares (OLS) regression using the `linearmodels` or `statsmodels` libraries.

```python
from linearmodels import OLS

df_wage = pd.read_csv("wage.csv")
df_wage["intercept"] = 1 # add constant for intercept

# Regress wage on education and experience
model = OLS(df_wage["wage"], df_wage[["intercept", "educ", "exper"]])
results = model.fit()
print(results.summary)
```
